# convertor

```python
def rename(name):
    if name=="Ghaziabad":
        return "G.Z.B"
    elif name=="New York":
        return "U.S.A"
    else:
        return name
```

```python
df =pd.read_csv("convertor.csv",converters={"location":rename})
df
```
