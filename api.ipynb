```python
import requests
import pandas as pd
import time
```

```python
product_id=[]
title=[]
description=[]
price =[]
discountPercentage=[]
rating=[]
stock=[]
brand=[]
category=[]
```

```python
session = requests.Session()
def fetchApi(productId):
    url = f"https://api.freeapi.app/api/v1/public/randomproducts/{productId}"
    try:
         
         response = session.get(url, timeout=5)
         data = response.json()
         if(data["success"] and "data" in data):
             user_data=data["data"]
             # Now start filling the data.....
             product_id.append(user_data["id"])
             title.append(user_data["title"])
             description.append(user_data["description"])
             price.append(user_data["price"])
             discountPercentage.append(user_data["discountPercentage"])
             rating.append(user_data["rating"])
             stock.append(user_data["stock"])
             brand.append(user_data["brand"])
             category.append(user_data["category"])
     
    except:
        print("Error !!! While fetching data.")
    
for i in range(1,101):
    fetchApi(i)
    time.sleep(0.5)
```

```python
data = {"_id":product_id,"title":title,"description":description,"price":price,
        "discountPercentage":discountPercentage,"rating":rating,"stock":stock,
        "brand":brand,"category":category}
df = pd.DataFrame(data)
```

```python
print(df.head(10))
df.to_csv("products.csv", index=False)
```
