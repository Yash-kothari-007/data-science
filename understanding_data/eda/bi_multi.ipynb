```python
sns.boxplot(x="Sex",y="Age",hue="Survived",data=titanic[titanic["Age"].between(60,70)])
```

# categorical-categorical

```python
pd.crosstab(titanic["Sex"], titanic["Survived"])
```

```python
sns.heatmap(pd.crosstab(titanic["Sex"], titanic["Survived"]))
```

```python
#titanic.groupby('Pclass').mean()["Survived"]*100    P class is non numeric
titanic.groupby('Pclass')["Survived"].mean()*100
```

```python
print(pd.crosstab(titanic["SibSp"],titanic["Survived"]))
sns.clustermap(pd.crosstab(titanic["SibSp"],titanic["Survived"]))
```
