```python
import pandas as pd
import requests
import time
```

```python
products=[]
session = requests.Session()
def fetch(productId):
    url=f"https://api.freeapi.app/api/v1/public/randomproducts/{productId}"
    try:
        response = session.get(url, timeout=10)
        data = response.json()
        if "data" in data and data["success"]:
            products.append(data["data"])
    except Exception as e:
        print(f"Failed to fetch {productId}: {e}")
for i in range(1,101):
   fetch(i)
   time.sleep(0.5)
```

```python
df = pd.DataFrame(products)
df.drop(columns=["description","thumbnail","images"],inplace=True)
df.to_csv("apiData.csv",index=False)
```
